# Consolidando informações da COVID-19 para o estado de SC

## Vamos usar a API do brasil.IO

[Documentação da API](https://github.com/turicas/covid19-br/blob/master/api.md)

```python
import pandas as pd
from urllib.request import urlopen 
import json
```

```python
url_sc = "https://brasil.io/api/dataset/covid19/caso/data/?state=SC&place_type=state"
```

Vamos pegar os dados mais recentes da API e converter em um objeto de `python`

```python
with urlopen(url_sc) as request:
    resposta = request.read().decode()
    resposta_json = json.loads(resposta)
```

Nossa `resposta_json` contém a chave `results` que é onde vamos extrair os dados

```python
df = pd.DataFrame(resposta_json['results'])
```
